```python
import pyarrow.parquet as pq
import pandas as pd
import numpy as np
import os
```

```python
table = pq.read_table("raw data/gadm1_era_gust_un__daily.parquet")
df = table.to_pandas()
```

```python
df["Date"] = pd.to_datetime(df["Date"].str.replace("^X", "", regex=True),
                             format="%Y%m%d")

long_df = (
    df
    .melt(
        id_vars="Date",
        var_name="GID_1",
        value_name="maxwind"
    )
)

long_df["year"] = long_df["Date"].dt.year

annual_max = (
    long_df
    .groupby(["GID_1", "year"], as_index=False)["maxwind"]
    .max()
)

annual_max["GID_1"] = annual_max["GID_1"].str.replace("_", ".", n=1)
```

```python
annual_max.to_csv("processed data/maxgust_gadm1_era5.csv", index=False)
```
